```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## Гистограмма

```python
rng = np.random.default_rng(1)
data = rng.normal(size=1000)

plt.hist(data)
```

```python
plt.hist(data, bins=30, density=True, alpha=0.5)
```

```python
plt.hist(data, bins=30, density=True, histtype="step", edgecolor="red")
```

```python
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 1000)
x3 = rng.normal(3, 2, 1000)

args = dict(alpha=0.3, bins=40)

plt.hist(x1, **args)
plt.hist(x2, **args)
plt.hist(x3, **args)
```

```python
print(np.histogram(x1, bins=1))
print(np.histogram(x1, bins=2))
print(np.histogram(x1, bins=40))
```

## Двумерные гистограммы

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]

x, y = rng.multivariate_normal(mean, cov, 1000).T

plt.hist2d(x, y, bins=30)
cb = plt.colorbar()
cb.set_label("Point in interval")
```

```python
print(np.histogram2d(x, y, bins=1))
print(np.histogram2d(x, y, bins=10))
```

```python
plt.hexbin(x, y, gridsize=30)
```

## Легенда

```python
x = np.linspace(0, 10, 1000)

fig, ax = plt.subplots()
ax.plot(x, np.sin(x), label="Синус")
ax.plot(x, np.cos(x), label="Косинус")
ax.axis("equal")
ax.legend()
fig.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), label="Синус")
ax.plot(x, np.cos(x), label="Косинус")
ax.axis("equal")
ax.legend(frameon=False)
fig.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), label="Синус")
ax.plot(x, np.cos(x), label="Косинус")
ax.plot(x, np.cos(x) + 2, label="Косинус")
ax.axis("equal")
ax.legend(frameon=True, fancybox=True, shadow=True)
fig.show()
```

```python
y = np.sin(x[:, np.newaxis] + np.pi * np.arange(0, 2, 0.5))
lines = plt.plot(x, y)

plt.legend(lines, ['1', "второй", "third", "4-ый"], loc="upper left")
```

```python
lines = plt.plot(x, y)
plt.legend(lines[:2], ['1', '2'])
```

```python
cities = pd.read_csv('./data/california_cities.csv')
lat, lon, pop, area = cities['latd'], cities['longd'],\
    cities['population_total'], cities['area_total_km2']

plt.scatter(lon, lat, c=pop, s=area)
```

```python
plt.scatter(lon, lat, c="red", s=10)
```

```python
plt.scatter(lon, lat, c=np.log10(pop), s=area)
plt.xlabel("Широта")
plt.ylabel("Долгота")
plt.colorbar()
plt.clim(3, 7)

plt.scatter([], [], c='k', alpha=0.5, s=100, label="100 $km^2$")
plt.scatter([], [], c='k', alpha=0.5, s=300, label="300 $km^2$")
plt.scatter([], [], c='k', alpha=0.5, s=500, label="500 $km^2$")
plt.legend(labelspacing=3, frameon=False)
```

```python
fig, ax = plt.subplots()

lines = []
styles = ['-', '--', '-.', ':']
for i in range(4):
    lines += ax.plot(x, np.sin(x - i + np.pi/2), styles[i])
ax.axis("equal")
ax.legend(lines[:2], ["line 1", "line 2"], loc="upper right")
leg = mpl.legend.Legend(ax, lines[1:], ['line 2', 'line 3', 'line 4'], loc='lower left')
ax.add_artist(leg)
fig.show()
```

## Простые сетки

```python
x = np.linspace(0, 10, 1000)

fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.plot(np.sin(x + np.pi / 4 * 4))
fig.show()
```
