```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(8, 5))

X = np.array([1.0, 5.0, 10.0, 15.0, 20.0])
red_Y = np.array([0.5, 7.0, 3.0, 5.0, 11.0])
green_Y = np.array([4.0, 3.0, 0.5, 8.0, 12.0])

plt.plot(X, red_Y, '-or', label='line1')
plt.plot(X, green_Y, '-.og', label='line1')
plt.legend()
```

```python
fig = plt.figure(figsize=(8, 4))
grid = plt.GridSpec(2, 4)

upper = fig.add_subplot(grid[0, :])
left = fig.add_subplot(grid[1, :2])
right = fig.add_subplot(grid[1, 2:])

upper.plot(np.arange(1, 6), np.array([0.5, 7., 6., 3., 5.]))
left.plot(np.arange(1, 6), np.array([9., 4., 2., 4., 9.]))
right.plot(np.arange(1, 6), np.array([-7., -4., 2., -4., -6.]))
```

```python
fig = plt.figure(figsize=(8, 5))
plt.plot(np.arange(-5, 6), np.array([25, 16, 9, 4, 1, 0, 1, 4, 9, 16, 25]))
plt.annotate('min', xy=(0, 0), xytext=(0, 10), arrowprops={'facecolor': 'green'})
```

```python
data = np.random.randint(0, 11, (7, 7))
plt.imshow(data, cmap='viridis')
plt.colorbar()
```

```python
x = np.linspace(0, 5, 100)
y = np.cos(3 * x)

plt.fill_between(x, y, 0, where=(y > 0), color='blue', alpha=0.6)  #  
plt.fill_between(x, y, 0, where=(y < 0), color='blue', alpha=0.6)
plt.plot(x, y, color='red')  #   
```

```python
x = np.linspace(0, 5, 100)
y = np.cos(3 * x)

y_masked = np.ma.masked_where(y < 0, y)  

plt.plot(x, y_masked, linewidth=2)
```

```python
fig, ax = plt.subplots(1, 3)

ax[0].set_aspect(1)
ax[0].grid()

ax[1].set_aspect(1)
ax[1].grid()

ax[2].set_aspect(1)
ax[2].grid()
for i in range(6):
    ax[0].plot([i, i], [i, i + 1], color='green')
    ax[0].plot([i, i + 1], [i + 1, i + 1], color='green')
```
